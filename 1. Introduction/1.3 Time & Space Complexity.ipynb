# Time Complexity

- Time complexity in computer science describes how the execution time of an algorithm grows as the size of input increases.
- It essentially quantifies the number of operations an algorithm performs as a function of the input data size.

# Asymptotic Notations

## 📌 Introduction
Asymptotic notations are used in computer science to describe the **time and space complexity** of algorithms. They help us understand how an algorithm performs as the input size grows.

---

## 🚀 Types of Asymptotic Notations

### 1. Big-O Notation (O)
- Represents the **upper bound** (worst-case scenario)
- Describes the maximum time an algorithm can take

**Example:**
O(n), O(n²), O(log n)

---

### 2. Big-Omega Notation (Ω)
- Represents the **lower bound** (best-case scenario)
- Describes the minimum time required

**Example:**
Ω(1), Ω(n)

---

### 3. Big-Theta Notation (Θ)
- Represents the **tight bound** (average case)
- Gives both upper and lower bounds

**Example:**
Θ(n), Θ(n log n)

---

## 📊 Common Complexities

| Complexity | Name           | Example Use Case        |
|------------|----------------|------------------------|
| O(1)       | Constant       | Array access           |
| O(log n)   | Logarithmic    | Binary search          |
| O(n)       | Linear         | Loop over elements     |
| O(n log n) | Linearithmic   | Merge sort             |
| O(n²)      | Quadratic      | Nested loops           |

---

## 🎯 Why It Matters
- Helps compare algorithms
- Predicts performance for large inputs
- Guides optimization and efficient coding

---

## ✅ Conclusion
Asymptotic notations provide a simple way to analyze and compare algorithms, making them essential for writing efficient programs.

# Types of Time Complexities

Time complexities are categorized based on how an algorithm's runtime scales with the input size.
Common types include constant time (O(1)), Logarithmic time (O(log n)), Linear time (O(n)), Quadratic time (O(n^2)), and Exponential time (O(2^n))
