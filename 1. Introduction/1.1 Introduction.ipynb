# What is a Programming Language ?

A programming language is a formal language that provides instructions for a computer to execute. It's how humans communicate with computers to create software, applications, and other technology.

# Low-Level vs High-Level Programming Languages

## Low-Level Programming Languages
- Close to machine (hardware) level  
- Includes Machine Language and Assembly Language  
- Hard to read and write  
- Faster execution and more control over hardware  
- Example: Assembly  

### Pros
- High performance  
- Direct hardware control  

### Cons
- Difficult to learn  
- Not portable  

---

## High-Level Programming Languages
- Closer to human language  
- Easy to read, write, and maintain  
- Requires a compiler/interpreter  
- Slower compared to low-level languages  

- Examples: Python, Java, C++, JavaScript  

### Pros
- Easy to learn and use  
- Portable across platforms  
- Faster development  

### Cons
- Less control over hardware  
- Slightly lower performance  

---

## Key Differences

| Feature        | Low-Level Language | High-Level Language |
|----------------|------------------|--------------------|
| Abstraction    | Low              | High               |
| Ease of Use    | Difficult        | Easy               |
| Speed          | Fast             | Moderate           |
| Portability    | Low              | High               |
| Control        | High             | Limited            |

# Compiler vs Interpreter

## Compiler
- Translates entire program at once  
- Generates machine code (executable file)  
- Errors shown after full compilation  
- Faster execution  

### Examples
- C, C++

### Pros
- High performance  
- Optimized code  

### Cons
- Compilation takes time  
- Debugging is harder  

---

## Interpreter
- Translates code line by line  
- No separate executable file  
- Stops at first error  
- Slower execution  

### Examples
- Python, JavaScript  

### Pros
- Easy debugging  
- Faster development  

### Cons
- Slower execution  
- Requires interpreter every time  

---

## Key Differences

| Feature        | Compiler              | Interpreter          |
|----------------|----------------------|----------------------|
| Translation    | Whole program        | Line by line         |
| Output         | Executable file      | No executable        |
| Speed          | Fast                 | Slower               |
| Error Handling | After compilation    | Immediate            |
| Examples       | C, C++               | Python, JavaScript   |

# Introduction to Python

## What is Python?
Python is a high-level, interpreted, and general-purpose programming language known for its simplicity and readability.

## Features of Python
- Easy to learn and use  
- Simple and readable syntax  
- Interpreted language  
- Cross-platform (portable)  
- Supports multiple programming paradigms (OOP, functional, procedural)  
- Large standard library  

## Applications of Python
- Web Development (Django, Flask)  
- Data Science & Machine Learning  
- Automation & Scripting  
- Game Development  
- Desktop Applications
