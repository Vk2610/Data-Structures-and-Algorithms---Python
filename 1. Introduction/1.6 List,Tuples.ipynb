# Introduction to Lists

- In python, List allows us to store multiple items in a single variable.
- For example, if you need to store the ages of all students in a class, you can use a list to store all the ages together.
- Lists are similar to arrays (dynamic arrays that allows us to store multiple items in a single variable) in other programming languages, but they are more flexible and can contain different types of data.

```python
ages = [20, 22, 19, 25, 21]
print(type(ages))  
print(ages[2])

# with negative indexing
print(ages[-1])

# updating list 
ages[1] = 23
print(ages)
```

```python
# Heterogeneous list
my_list = [1, "Hello", 3.14, True]
print(my_list)
print(type(my_list))

# length of list
print(len(my_list))
```

```python
# Extend vs Append
my_list.append("World")
print(my_list)

my_list.extend([1, 2, 3])
print(my_list)

my_list.insert(0, "Hello")
print(my_list)
```

```python
# Operations on List

# take a example list 
numbers = [1, 2, 3, 4, 5]
# sum of list
print(sum(numbers))

# max and min
print(max(numbers))

print(min(numbers))

# sorting a list
numbers.sort()

print(numbers)

# reverse a list
numbers.reverse()

print(numbers)
```

```python
# List functions

# count occurrences
print(numbers.count(3))

# index of an element
print(numbers.index(4))

# remove an element
numbers.remove(2)

print(numbers)
```
