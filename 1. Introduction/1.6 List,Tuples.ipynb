# Introduction to Lists

- In python, List allows us to store multiple items in a single variable.
- For example, if you need to store the ages of all students in a class, you can use a list to store all the ages together.
- Lists are similar to arrays (dynamic arrays that allows us to store multiple items in a single variable) in other programming languages, but they are more flexible and can contain different types of data.

```python
ages = [20, 22, 19, 25, 21]
print(type(ages))  
print(ages[2])

# with negative indexing
print(ages[-1])

# updating list 
ages[1] = 23
print(ages)
```

```python
# Heterogeneous list
my_list = [1, "Hello", 3.14, True]
print(my_list)
print(type(my_list))

# length of list
print(len(my_list))
```

```python
# Extend vs Append
my_list.append("World")
print(my_list)

my_list.extend([1, 2, 3])
print(my_list)

my_list.insert(0, "Hello")
print(my_list)
```

```python
# Operations on List

# take a example list 
numbers = [1, 2, 3, 4, 5]
# sum of list
print(sum(numbers))

# max and min
print(max(numbers))

print(min(numbers))

# sorting a list
numbers.sort()

print(numbers)

# reverse a list
numbers.reverse()

print(numbers)
```

```python
# List functions

# count occurrences
print(numbers.count(3))

# index of an element
print(numbers.index(4))

# remove an element
numbers.remove(2)

print(numbers)
```

```python
# Deep vs Shallow Copy
original_list = [1, 2, 3, [4, 5]]
shallow_copy = original_list.copy()
deep_copy = original_list[:]

print("Original List:", original_list)
print("Shallow Copy:", shallow_copy)
print("Deep Copy:", deep_copy)
# Modifying the nested list in original_list
original_list[3][0] = 'Changed'

print("Original List after modification:", original_list)
print("Shallow Copy after modification:", shallow_copy)
print("Deep Copy after modification:", deep_copy) 
```

# Tuples

- A tuple is a collection of ordered, immutable (unchangeable) elements.
- Tuples are defined using parentheses `()` and can contain elements of different data types.
- Once a tuple is created, its elements cannot be modified, added, or removed. This makes tuples useful for storing data that should not be changed throughout the program.

```python
# Example tuple 

my_tuple = (1, "Hello", 3.14, True)
print(my_tuple)
print(type(my_tuple))
```

```python
# Tuple operations  

# length of tuple
print(len(my_tuple))

# accessing elements
print(my_tuple[1])

# tuple unpacking
a, b, c, d = my_tuple
print(a)
print(b)
print(c)
print(d)

# tuple functions
print(my_tuple.count(1))
print(my_tuple.index(3.14))

# tuple methods
print(my_tuple.index(3.14))
print(my_tuple.count(1))

# tuple to list
my_list = list(my_tuple)
print(my_list)
print(type(my_list))

# list to tuple
my_tuple = tuple(my_list)
print(my_tuple)
print(type(my_tuple))
```

# String

- A string is a sequence of characters enclosed in single quotes `' '`, double quotes `" "`, or triple quotes `''' '''` or `""" """`.
- Strings are used to represent text data in Python and can contain letters, numbers, symbols, and whitespace.
- Strings are immutable, meaning that once a string is created, it cannot be modified. However, you can create new strings by concatenating or slicing existing strings.

```python
# example string
my_string = "Hello, World!"
print(my_string)
print(type(my_string))
```

```python
# Operations on strings

# length of string
print(len(my_string))

# accessing characters
print(my_string[7])

# slicing strings
print(my_string[0:5])  # Hello

# string methods
print(my_string.upper())
print(my_string.lower())
print(my_string.replace("World", "Universe"))
print(my_string.split(" "))


# string formatting
name = "Alice"
age = 30
formatted_string = f"My name is {name} and I am {age} years old."
print(formatted_string)
```

```python
# Q1. Reverse a string
def reverse_string(s):
    return s[::-1]
print(reverse_string("Hello, World!"))
```

```python
# Q2. Check if a string is a palindrome
def is_palindrome(s):
    return s == s[::-1]
print(is_palindrome("madam"))
print(is_palindrome("hello"))
```

```python
# Q3 Find the largest number in a list
def find_largest(numbers):
    if not numbers:
        return None
    largest = numbers[0]
    for num in numbers:
        if num > largest:
            largest = num
    return largest
  
print(find_largest([3, 1, 4, 1, 5, 9]))
```

```python
# Q4. Count the occurrences of an element in a list
def count_occurrences(lst, element):
    count = 0
    for item in lst:
        if item == element:
            count += 1
    return count
  
print(count_occurrences([1, 2, 3, 2, 4, 2], 2))
```

```python
# Q1. Running sum problem 
# description : We have to find a sum of all the elements in a list and return a new list where each element at index i is the sum of elements from index 0 to i in the original list.
def running_sum(nums):
    ans = []
    ans.append(nums[0])
    
    for i in range(1, len(nums)):
        ans.append(ans[i-1] + nums[i])
        
    return ans
  
# Test 
print(running_sum([1, 2, 3, 4]))  # Output: [1, 3, 6, 10]
```

```python
# Q2. Remove duplicates from a sorted list

class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        n = len(nums)

        # If there are 2 or fewer elements, return the length
        if n <= 2:
            return n

        # First two elements are always allowed
        k = 2

        # Traverse from the third element
        for i in range(2, n):
            # Keep the element only if it doesn't create three duplicates
            if nums[i] != nums[k - 2]:
                nums[k] = nums[i]
                k += 1

        return k
      
# Test
solution = Solution()
nums = [1, 1, 1, 2, 2, 3]
new_length = solution.removeDuplicates(nums)
print(new_length)  # Output: 5
print(nums[:new_length])  # Output: [1, 1, 2, 2, 3]
```

```python
# Q3. Find the maximum product of two integers in a list
def max_product(nums):
    nums.sort()
    return max(nums[0] * nums[1], nums[-1] * nums[-2])
  
print(max_product([1, 2, 3, 4, 5]))
```

```python
# Q4. Find the first non-repeating character in a string
def first_non_repeating_char(s):
    char_count = {}
    for char in s:
        char_count[char] = char_count.get(char, 0) + 1
    for char in s:
        if char_count[char] == 1:
            return char
    return None
  
print(first_non_repeating_char("abaccdeff"))
```

```python
# Q4. Maximum subarray
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        curr_sum = 0
        max_sum = nums[0]

        for i in range(len(nums)):
            curr_sum += nums[i]

            if curr_sum > max_sum:
                max_sum = curr_sum

            if curr_sum < 0:
                curr_sum = 0

        return max_sum
      
# Test
solution = Solution()
solution.maxSubArray([-2,1,-3,4,-1,2,1,-5,4])
```

```python
# Matrix problem 
# description : You are given an m x n integer matrix matrix with the following two properties:
# Each row is sorted in non-decreasing order.

# solution

class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        m, n = len(matrix), len(matrix[0])
        row, col = 0, n - 1

        while row < m and col >= 0:
            if matrix[row][col] == target:
                return True
            elif matrix[row][col] < target:
                row += 1
            else:
                col -= 1

        return False
      
# Test
solution = Solution()
matrix = [[1,4,7,11,15],[2,5,8,12,19],[3,6,9,16,22],[10,13,14,17,24],[18,21,23,26,30]]
target = 5
print(solution.searchMatrix(matrix, target))  # Output: True
```

```python
# 121 : Buy or sell stock - Part 1
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        min_price = prices[0]
        profit=0

        for i in range(1,len(prices)):
            curr_profit = prices[i] - min_price
            if curr_profit > profit:
                profit=curr_profit
            min_price = min(min_price,prices[i])

        return profit
      
prices = [7,1,5,3,6,4]
solution = Solution()
print(solution.maxProfit(prices))
```

```python
# 1672 - Richest customer wealth 
class Solution:
    def maximumWealth(self, accounts: List[List[int]]) -> int:
        max_wealth = 0

        for customer in accounts:
            wealth = sum(customer)
            max_wealth = max(max_wealth, wealth)

        return max_wealth
```
